```python
import random

class BMICalculator:
    def __init__(self):
        self.invalid_messages = [
            "Oops! That doesn't look like a valid choice. Try again!",
            "Hmm, that's not one of the options. Please select a valid choice.",
            "Invalid input. Let's give it another shot!",
            "That doesn't seem right. Please enter a valid option.",
            "Whoops! Looks like you missed the mark. Let's try again.",
            "Invalid choice. Please choose from the given options.",
            "I didn't quite get that. Pick one of the listed options, please.",
            "Not quite what I expected! Give it another go.",
            "Hmm, I can't work with that. Please enter a valid choice."
        ]
        self.exit_messages = [
            "Actually, you know what? Maybe today isn't the right time to calculate your BMI. That's perfectly okay. You can always come back whenever you're ready.",
            "You know, I was thinking, let's not bother with it today. Take your time, and you can come back whenever you feel like checking your BMI.",
            "On second thought, maybe it's best to skip the BMI for now. No worries! Whenever you're in the mood, we can try again.",
            "Well, I’ve changed my mind. Today doesn’t seem like the day for BMI. But don’t worry, you can always come back whenever you’re ready!",
            "You know what, forget it for now. You don't seem too keen on checking your BMI today. Feel free to return whenever you want.",
            "Actually, never mind. It’s fine if you don’t want to check your BMI today. Come back whenever you feel like it!",
            "You know what, I've reconsidered. Let's put this off for today. You can always come back and check your BMI when you're ready.",
            "I was going to ask you for more details, but let’s leave it for another time. You can return anytime you want to calculate your BMI.",
            "Actually, I think it's better to skip it today. No problem at all, feel free to come back whenever you're ready for your BMI.",
            "I’ve had a change of heart. It’s okay to not check your BMI today. Come back when you're in the mood for it."
        ]
        self.menu()

    def random_invalid_message(self):
        print(random.choice(self.invalid_messages))

    def random_exit_message(self):
        print(random.choice(self.exit_messages))

    def menu(self):
        user_input = input(
            "Welcome! Your BMI (Body Mass Index) helps assess your weight in relation to height.\n"
            "Knowing your BMI is crucial for understanding health risks and managing your body better!\n"
            "Would you like to know your Body Mass Index today?\n"
            "1. Go ahead.\n"
            "2. Exit.\n"
        )
        if user_input == "1":
            self.get_height()
        elif user_input == "2":
            print("No problem! You can come back again anytime.")
            return
        else:
            self.random_invalid_message()
            self.menu2()

    def menu2(self):
        attempts = 0
        while attempts < 2:
            user_input2 = input(
                "Should we try and find out your BMI?\n"
                "1. Go ahead.\n"
                "2. Exit.\n"
            )
            if user_input2 == "1":
                self.cal_bmi()
                return
            elif user_input2 == "2":
                print("No problem! You can come back again anytime.")
                return
            else:
                self.random_invalid_message()
                attempts += 1
        self.random_exit_message()

    def get_height(self):
        attempts = 0
        while attempts < 2:
            height_unit = input(
                "Alright then! We'll need your measurements first. Let's start with height. Please choose the unit you want to use to enter your height in:\n"
                "1: ft., in.\n"
                "2: m, cm.\n"
                "3: Nevermind. I'll do this some other time.\n"
            )
            if height_unit == "1":
                attempts = 0
                while attempts < 2:
                    height_input = input(
                        "I see you've chosen feet and inches. That's cool, we can work with that.\n"
                        "Enter your height in decimal digits.\n"
                        "(If you're 5 ft 7 in tall, enter it as 5.7):\n"
                        "Your height: "
                    )
                    try:
                        height = float(height_input)
                        if height > 0:
                            self.height = height * 0.3048
                            self.get_weight()
                            return
                        else:
                            print("Height must be a positive number. Please try again.")
                    except ValueError:
                        print("Invalid input. Let's try that again.")
                        attempts += 1
                self.random_exit_message()
                return
            elif height_unit == "2":
                attempts = 0
                while attempts < 2:
                    height_input = input(
                        "I see you've chosen meters and centimeters.\n"
                        "Enter your height in meters (e.g., 1.75):\n"
                        "Your height: "
                    )
                    try:
                        height = float(height_input)
                        if height > 0:
                            self.height = height
                            self.get_weight()
                            return
                        else:
                            print("Height must be a positive number. Please try again.")
                    except ValueError:
                        print("Invalid input. Let's try that again.")
                        attempts += 1
                self.random_exit_message()
                return
            elif height_unit == "3":
                print("That's alright. You can come back when you're ready.")
                self.random_exit_message()
                return
            else:
                self.random_invalid_message()
                attempts += 1
        self.random_exit_message()
        return

    def get_weight(self):
        attempts = 0
        while attempts < 2:
            weight_unit = input(
                "Great, we have your height! Next, we'll need your weight.\n"
                "What unit do you want to use for your weight?:\n"
                "1: lbs, oz.\n"
                "2: kgs, cms\n"
                "3: Nevermind. I'll do this some other time.\n"
            )
            if weight_unit == "1":
                attempts = 0
                while attempts < 2:
                    weight_input = input(
                        "Ah, pounds and ounces, it is! That's cool, we can work with that.\n"
                        "Enter your weight in decimal digits:\n"
                        "Your weight: "
                    )
                    try:
                        weight = float(weight_input)
                        if weight > 0:
                            self.weight = weight * 0.453592
                            self.cal_bmi()
                            return
                        else:
                            print("Weight must be a positive number. Please try again.")
                    except ValueError:
                        print("Invalid input. Let's try that again.")
                        attempts += 1
                self.random_exit_message()
                return
            elif weight_unit == "2":
                attempts = 0
                while attempts < 2:
                    weight_input = input(
                        "As you wish!.\n"
                        "Enter your weight in kilograms (e.g., 60.5):\n"
                        "Your weight: "
                    )
                    try:
                        weight = float(weight_input)
                        if weight > 0:
                            self.weight = weight
                            self.cal_bmi()
                            return
                        else:
                            print("Weight must be a positive number. Please try again.")
                    except ValueError:
                        print("Invalid input. Let's try that again.")
                        attempts += 1
                self.random_exit_message()
                return
            elif weight_unit == "3":
                print("That's alright. You can come back when you're ready.")
                self.random_exit_message()
                return
            else:
                self.random_invalid_message()
                attempts += 1
        self.random_exit_message()
        return

    def cal_bmi(self):
        bmi = self.weight / (self.height ** 2)
        if bmi < 18.5:
            category = "underweight"
        elif 18.5 <= bmi < 24.9:
            category = "normal weight"
        elif 25 <= bmi < 29.9:
            category = "overweight"
        else:
            category = "obese"
    
        # Display the BMI and category
        print(f"\nYour BMI is {bmi:.2f}. Based on the BMI scale, you would be classified as {category}.\n")
        print("Thank you for using the BMI Calculator!")


bmi_calculator = BMICalculator()
```
